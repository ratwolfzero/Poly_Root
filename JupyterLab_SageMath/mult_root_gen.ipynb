```python
x = var('x')

n = 7          # number of roots
m = 9          # multiplicity of each root

roots = [i for i in range(1, n+1)]   # e.g. [1,2,3,4]

f = prod((x - r)^m for r in roots)

# expand and get coefficients
coeffs = f.coefficients(sparse=False)
coeffs = list(reversed(coeffs))

print(" ".join(str(c) for c in coeffs))
```
